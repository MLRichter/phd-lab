```python
df = pd.read_csv("./logs/MPNet36_1_4_3_7/Cifar10_32/DIS/gradient_norms.csv", index_col=0)
short = df[df["name"].apply(lambda x: "pathway0" in x)]
long =  df[df["name"].apply(lambda x: "pathway1" in x)]
```

```python
k = list(range(len(short["grad_in"].values)))
k = [i*4 for i in k]
plt.plot(k, short["grad_in"].values[::-1], label="short")
plt.plot(list(range(len(long["grad_in"].values))), long["grad_in"].values[::-1])
plt.scatter(k, short["grad_in"].values[::-1])
plt.scatter(list(range(len(long["grad_in"].values))), long["grad_in"].values[::-1], label="long")
plt.legend()
```

```python
plt.figure(figsize=(10, 3))
plt.plot(k, short["grad_out"].values[::-1], label="short path layers")
plt.plot(list(range(len(long["grad_out"].values))), long["grad_out"].values[::-1], label="long path layers")
plt.scatter(k, short["grad_out"].values[::-1])
plt.scatter(list(range(len(long["grad_out"].values))), long["grad_out"].values[::-1])
plt.legend()
plt.xlabel("Layers")
plt.ylabel("Mean Absolute Gradient")
plt.grid()
plt.savefig("vanishing_gradient_MPNet36_1_4_3_7.pdf")
```
